```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

### Loading Df

```python
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing()

housing_df = pd.DataFrame(housing.data, columns=housing.feature_names)
housing_df["target"] = housing["target"]
```

```python
housing_df
```

```python
from sklearn.model_selection import train_test_split

# Setup random seed
np.random.seed(42)

# Create the data
X = housing_df.drop("target", axis=1)
y = housing_df["target"]  # median house price in $100,00s

# Bin counts for stratifying
count_bins = pd.qcut(y, q=10, labels=False, duplicates="drop")

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=count_bins)
```

### Ridge (out of the box, default parameters, raw data)

```python
from sklearn.linear_model import Ridge

# Instantiate and fit the model (on the training set)
ridge_base = Ridge()
ridge_base.fit(X_train, y_train)

# Check the score of the model (on the test set)
ridge_base.score(X_test, y_test)
```

### Lasso (out of the box, default parameters, raw data)

```python
from sklearn.linear_model import Lasso

lasso_base = Lasso()
lasso_base.fit(X_train, y_train)

lasso_base.score(X_test, y_test)
```

### SVM-linear (out of the box, default parameters, raw data)

```python
from sklearn.svm import SVR

svr_base = SVR(kernel="linear")
svr_base.fit(X_train, y_train)

svr_base.score(X_test, y_test)
```
