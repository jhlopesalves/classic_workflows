# Ames Housing Dataset


## Ames Housing (Ames, Iowa) — Dataset Overview and Feature Map

The **Ames Housing** dataset is a widely used real-estate regression benchmark describing residential property sales in **Ames, Iowa**. It is valued in teaching and modelling because it combines **rich, human-interpretable housing attributes** with a realistic mix of **numeric, ordinal, and nominal** predictors.

> Important naming note  
> “Ames Housing” may refer to **(a)** Dean De Cock’s full curated dataset (2930 observations, sales from 2006–2010), or **(b)** the popular **Kaggle/OpenML** variant used in many machine-learning tutorials (typically **1460 rows** with `SalePrice` available for supervised learning). The feature definitions are largely consistent, but **row counts and packaging differ**.

### Core facts

| Property | Value |
|---|---|
| **Original provenance** | Ames City Assessor’s Office records (Ames, Iowa), curated for education and regression modelling |
| **Reference publication** | De Cock (2011), *Journal of Statistics Education* |
| **Common ML distribution** | OpenML `data_id=42165` (often accessed via `sklearn.datasets.fetch_openml`) and the Kaggle “House Prices” competition packaging |
| **Target variable** | `SalePrice` (sale price; USD) |
| **Predictors** | **79 explanatory variables** (plus identifiers such as `Id`) |
| **Data types (original curation)** | A mix of nominal, ordinal, discrete, and continuous variables |
| **Time span (original curation)** | Sales from **2006–2010** |


## Feature Categories (high-level map)

Below is a **structured map** of the variables commonly used in the Kaggle/OpenML packaging. For full definitions and level encodings, use the official data dictionary (see References).

### Preprocessing Pipeline

Building scikit-learn transformers to handle imputation, encoding, and scaling in a reproducible pipeline.

```python
from sklearn.base import BaseEstimator, TransformerMixin


# Custom transformer: converts year columns to age (years since reference)
# BaseEstimator: provides get_params/set_params for sklearn integration
# TransformerMixin: provides fit_transform from fit + transform
class YearToAgeTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, reference_year=2025):
        self.reference_year = reference_year

    def fit(self, X, y=None):
        return self  # Stateless transformer - nothing to learn

    def transform(self, X):
        return self.reference_year - X
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OrdinalEncoder

# Quality rating columns share the same ordinal scale (Po < Fa < TA < Gd < Ex)
ordinal_cols = [
    "ExterQual",
    "ExterCond",
    "BsmtQual",
    "BsmtCond",
    "HeatingQC",
    "KitchenQual",
    "FireplaceQu",
    "GarageQual",
    "GarageCond",
]

# OrdinalEncoder maps categories to integers preserving order
quality_order = ["None", "Po", "Fa", "TA", "Gd", "Ex"]  # 0 to 5

# Pipeline chains transformers: impute first, then encode
ordinal_pipe = Pipeline([
    ("impute", SimpleImputer(strategy="constant", fill_value="None")),
    ("encoder", OrdinalEncoder(categories=[quality_order] * len(ordinal_cols))),
])
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

# Categorical columns where NaN semantically means "None" (no feature present)
cat_absence_cols = [
    "GarageType",
    "GarageFinish",
    "BsmtFinType2",
    "BsmtExposure",
    "BsmtFinType1",
    "MasVnrType",
]

# OneHotEncoder: creates binary columns for each category
categorical_pipe = Pipeline([
    ("impute", SimpleImputer(strategy="constant", fill_value="None")),
    ("encoder", OneHotEncoder(handle_unknown="ignore", sparse_output=False)),
])
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# Numeric columns where NaN means zero (no masonry, no garage)
num_absence_cols = ["MasVnrArea", "GarageYrBlt"]

# StandardScaler: centers to mean=0 and scales to std=1 (z-score)
numerical_pipe = Pipeline([
    ("impute", SimpleImputer(strategy="constant", fill_value=0)),
    ("scaler", StandardScaler()),
])
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# LotFrontage: missing likely due to irregular lots, impute with median
median_impute_cols = ["LotFrontage"]
median_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
])
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

# Electrical has only 1 missing value - impute with mode (most common)
mode_impute_cols = ["Electrical"]

mode_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),  # mode imputation
    ("encoder", OneHotEncoder(handle_unknown="ignore", sparse_output=False)),
])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# Convert year to age for more intuitive feature interpretation
# Age tends to have linear relationship with depreciation
years_age_cols = ["YearBuilt", "YearRemodAdd"]
years_pipe = Pipeline([
    ("years", YearToAgeTransformer(reference_year=2025)),
    ("scaler", StandardScaler()),
])
```

```python
# Complete numeric columns - no missing values, just need scaling
num_cols = [
    "LotArea",
    "BsmtFinSF1",
    "BsmtFinSF2",
    "BsmtUnfSF",
    "TotalBsmtSF",
    "1stFlrSF",
    "2ndFlrSF",
    "GrLivArea",
    "BsmtFullBath",
    "BsmtHalfBath",
    "FullBath",
    "HalfBath",
    "BedroomAbvGr",
    "KitchenAbvGr",
    "TotRmsAbvGrd",
    "Fireplaces",
    "GarageCars",
    "GarageArea",
    "WoodDeckSF",
    "OpenPorchSF",
    "EnclosedPorch",
    "3SsnPorch",
    "ScreenPorch",
    "PoolArea",
    "MiscVal",
    "MoSold",
    "YrSold",
]

# Nominal categorical columns - no inherent order, use one-hot encoding
categorical = [
    "MSSubClass",
    "MSZoning",
    "Street",
    "LotShape",
    "LandContour",
    "Utilities",
    "LotConfig",
    "LandSlope",
    "Neighborhood",
    "Condition1",
    "Condition2",
    "BldgType",
    "HouseStyle",
    "RoofStyle",
    "RoofMatl",
    "Exterior1st",
    "Exterior2nd",
    "Foundation",
    "Heating",
    "CentralAir",
    "Functional",
    "PavedDrive",
    "SaleType",
    "SaleCondition",
]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

general_one_hot_encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
general_scaler = StandardScaler()


# ColumnTransformer applies different transformations to different column subsets
# Format: (name, transformer, columns)
transformer = ColumnTransformer(
    [
        ("ordinal_encoder", ordinal_pipe, ordinal_cols),
        ("cat_absence_imputer", categorical_pipe, cat_absence_cols),
        ("num_absence_imputer", numerical_pipe, num_absence_cols),
        ("median_imputer", median_pipe, median_impute_cols),
        ("mode_imputer", mode_pipe, mode_impute_cols),
        ("years_to_age", years_pipe, years_age_cols),        
        ("num_scaler", general_scaler, num_cols),
        ("cat_encoder", general_one_hot_encoder, categorical),
    ],
    remainder="drop",  # Columns not listed are removed
)
```

### Baseline Model

1. Fit a basic model (no tuning, no complexity)
2. Evaluate it properly
3. Establish a baseline score to beat later

```python
from sklearn.compose import TransformedTargetRegressor
from sklearn.linear_model import Ridge
from sklearn.pipeline import Pipeline

base_model = TransformedTargetRegressor(Ridge(), func=np.log1p, inverse_func=np.expm1)

baseline_pipeline = Pipeline([
    ("preprocessing", transformer),
    ("model", base_model),
])

baseline_pipeline.fit(X_train, y_train)
y_pred = baseline_pipeline.predict(X_test)
```

### Baseline metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)

print(f"R2 Score: {r2:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
```

```python
baseline_rmse = np.sqrt(mean_squared_error(y_test, [y_train.mean()] * len(y_test)))
print(f"Baseline RMSE: {baseline_rmse:.2f}")
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

sns.scatterplot(x=y_test, y=y_pred, ax=ax, alpha=0.6)

# Add diagonal line for perfect prediction
min_val = min(y_test.min(), y_pred.min())
max_val = max(y_test.max(), y_pred.max())
ax.plot(
    [min_val, max_val],
    [min_val, max_val],
    "r--",
    linewidth=2,
    label="Perfect Prediction",
)

ax.set(xlabel="Actual Values", ylabel="Predicted Values")
ax.set_title("Actual vs Predicted Values", fontsize=14, fontweight="bold")
ax.grid(True, alpha=0.4, linewidth=0.5, color="grey")
ax.set_axisbelow(True)
ax.legend()

plt.tight_layout()
plt.show()
```

### First candidate: RandomForestRegressor

```python
from sklearn.compose import TransformedTargetRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.pipeline import Pipeline

rf_model = TransformedTargetRegressor(
    regressor=RandomForestRegressor(random_state=42),
    func=np.log1p,
    inverse_func=np.expm1,
)

rf_pipe = Pipeline([
    ("preprocessing", transformer),
    ("rf_model", rf_model),
])

rf_pipe.fit(X_train, y_train)
y_pred = rf_pipe.predict(X_test)


r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)

print(f"R2 Score: {r2:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
```

### Second Candidate: HistGradientBoostingRegressor


```python
from sklearn.compose import TransformedTargetRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.pipeline import Pipeline

gb_model = TransformedTargetRegressor(
    regressor=GradientBoostingRegressor(random_state=42),
    func=np.log1p,
    inverse_func=np.expm1,
)

gb_pipe = Pipeline([
    ("preprocessing", transformer),
    ("gb_model", gb_model),
])

gb_pipe.fit(X_train, y_train)
y_pred = gb_pipe.predict(X_test)


r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)

print(f"R2 Score: {r2:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
```

### Fine tuning hyperparameters

```python
from sklearn.model_selection import RandomizedSearchCV

param_dist = {
    "gb_model__regressor__n_estimators": [200, 500, 1000],
    "gb_model__regressor__learning_rate": [0.01, 0.05, 0.1],
    "gb_model__regressor__max_depth": [3, 4, 5],
    "gb_model__regressor__subsample": [0.8, 1.0],
}


gb_search = RandomizedSearchCV(
    estimator=gb_pipe,
    param_distributions=param_dist,
    n_iter=20,  # try 20 random combinations
    cv=5,  # 5-fold cross-validation
    scoring="neg_root_mean_squared_error",
    random_state=42,
    n_jobs=-1,  # use all CPU cores
)

gb_search.fit(X_train, y_train)
```

```python
y_pred = gb_search.best_estimator_.predict(X_test)

r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)

print(f"R2 Score: {r2:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
```
