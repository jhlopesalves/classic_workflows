```python
from urllib.request import urlopen

from bs4 import BeautifulSoup
```

```python
# Read the stream into a variable
html_content = urlopen("http://pythonscraping.com/pages/page1.html").read()

# Pass the VARIABLE to both parsers
bs = BeautifulSoup(html_content, "html.parser")
bs_lxml = BeautifulSoup(html_content, "lxml")

print(bs.h1)
print(bs_lxml.h1)
```
